```python
# YOUR CODE HERE
plt.xlabel('Amount of Co2')

plt.xticks(range(fuel_econ['co2'].min()-4, fuel_econ['co2'].max()+ 1, 50))
plt.yticks(range(0, 300, 20))

plt.xlim(fuel_econ['co2'].min(), fuel_econ['co2'].max())

bins = np.arange(fuel_econ['co2'].min(), fuel_econ['co2'].max()+5, 10)

plt.hist(data = fuel_econ, x = 'co2', bins = bins,
         edgecolor='k', alpha=0.5, color='green')

plt.title('Co2 distribution')
# print(plt.ylabel)
# print(fuel_econ['co2'].max())
# print(fuel_econ['co2'].min())
plt.show()
```

```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['highway'], bins=10, edgecolor='black', color='lightblue')
plt.title('Highway distribution')
plt.xlabel('Highway Values')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(fuel_econ['city'], bins=50, edgecolor='black', color='blue')
plt.title('Fuel Efficiency')
plt.xlabel('City Fuel Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a scatter plot with a heatmap
plt.figure(figsize=(10, 6))
sb.heatmap(data=fuel_econ[['displ', 'co2']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.xlabel('Engine Displacement (L)')
plt.ylabel('CO2 Production (g/mi)')
plt.title('Correlation Heatmap: Engine Displacement vs. CO2 Production')
plt.show()
```
